```python
import os
import numpy as np

from astropy.io.fits import getdata
from astropy import wcs
from astropy.io import fits
from astropy import units as u
from astropy import constants as con
from astropy.coordinates import SkyCoord

import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib

import aplpy

import math
from scipy.interpolate import interp1d

#matplotlib.use('PDF')

from __future__ import print_function
from scipy.optimize import curve_fit

deg_to_rad = 0.0174533
rad_to_deg = 57.2958
```

```python
dist_sqaure, dispersion_function = np.genfromtxt('houde_data.txt', delimiter='\t',unpack=True)
```

```python
# Dispersion fucntion vs sqaure of distance 
fig = plt.figure(figsize=(12,4))
plt.xlabel(r'$ \mathrm{Distance}^2\/\/ \ell^2\/\/ [\mathrm{arcmin}^2]$')
plt.ylabel(r'$[1 - <\mathrm{cos}(\Delta \Phi)>]$')
plt.scatter(dist_sqaure, dispersion_function)
plt.show() 
```

```python
# Dispersion fucntion vs distance (without square)
dist = np.sqrt(dist_sqaure)
fig = plt.figure(figsize=(12,4))
plt.xlabel(r'$ \mathrm{Distance}\/\/ \ell\/\/ [\mathrm{arcmin}]$')
plt.ylabel(r'$[1 - <\mathrm{cos}(\Delta \Phi)>]$')
plt.scatter(dist, dispersion_function)
plt.show() 
```

```python


W = 0.07666 # acrmin --> 4.6 arcsec ; taken from Houde 2009
Delta_prime = 3.5 # acrmin ; taken from Houde 2009

def correlation_func(ell2, delta, PAdispersion4, a_2_prime):
    correlation = np.sqrt(2*np.pi)*PAdispersion4*((delta**3)/((delta**2+2*W**2)*Delta_prime))*(1-np.exp((-ell2)/(2*(delta**2+2*W**2))))+a_2_prime*(ell2)
    return correlation

parameters, covariance = curve_fit(correlation_func, dist_sqaure, dispersion_function, bounds=([0.121667, 0, 0], [1., 1., 1]))

fit_delta = parameters[0]
fit_PAdispersion4 = parameters[1]
fit_a_2_prime = parameters[2]

print("The sqaure of PA dispersion is fitted to be", fit_PAdispersion4)
print("delta is fitted to be", fit_delta, "arcmin")
print("a' is fitted to be", fit_a_2_prime, "arcmin^2")
#print("W is", W, "arcsec")

correlation_func_fit = correlation_func(dist_sqaure, fit_delta, fit_PAdispersion4, fit_a_2_prime)
fig = plt.figure(figsize=(12,4))
plt.xlabel(r'$ \mathrm{Distance}^2\/\/ \ell^2\/\/ [\mathrm{arcmin^2}]$')
plt.ylabel(r'$[1 - <\mathrm{cos}(\Delta \Phi)>]$')
plt.scatter(dist_sqaure, dispersion_function)
plt.plot(dist_sqaure, correlation_func_fit, '-')
plt.show() 
```

```python
def linear_fuc(ell2, a_2_prime, c):
    linear = a_2_prime*ell2 + c
    return linear

parameters, covariance = curve_fit(linear_fuc, dist_sqaure[10:], dispersion_function[10:])

fit_slope = parameters[0]
fit_intercept = parameters[1]

print("The slope is fitted to be", fit_slope)
print("The x-intercept is fitted to be", fit_intercept)

linear_fuc_fit = linear_fuc(dist_sqaure, fit_slope, fit_intercept)
fig = plt.figure(figsize=(12,4))
plt.xlabel(r'$ \mathrm{Distance}^2\/\/ \ell^2\/\/ [\mathrm{arcsec}]$')
plt.ylabel(r'$[1 - <\mathrm{cos}(\Delta \Phi)>]$')
plt.scatter(dist_sqaure, dispersion_function)
plt.plot(dist_sqaure, linear_fuc_fit, '-')
plt.show() 
```
